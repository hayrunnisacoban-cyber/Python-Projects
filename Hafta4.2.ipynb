```python
#4. bölüm 2. konu 4. soru
def line(length, character="#"):
    """Print a line of characters with the specified length."""
    print(character * length)

def triangle(size):
    """
    Draw a triangle of hashes that is as tall and wide as the size argument.
    Uses the line function to print each row.
    """
    for i in range(1, size + 1):
        line(i)

# Examples
triangle(6)
print()
triangle(3)
```

```python
#4. bölüm 2. konu 5. soru

def line(length, character="#"):
    print(character * length)

def shape(size, triangle_char, rect_height, rect_char):

    # Draw the triangle
    for i in range(1, size + 1):
        line(i, triangle_char)

    # Draw the rectangle
    for i in range(rect_height):
        line(size, rect_char)

# Examples
shape(5, "X", 3, "*")
print()
shape(2, "o", 4, "+")
print()
shape(3, ".", 0, ",")
```

```python
# 4. bölüm 3. konu sondan 3. soru
def length(lst):

    count = 0
    for _ in lst:
        count += 1
    return count
my_list = [1, 2, 3, 4, 5]
result = length(my_list)
print("The length is", result)

# the list given as an argument doesn't need to be stored in any variable
result = length([1, 1, 1, 1])
print("The length is", result)
```

```python
#4. bölüm 3. konu sondan 2. sorudef mean(numbers):
    if not numbers:  # Handle empty list case
        return 0.0

    total_sum = 0
    count = 0

    for num in numbers:
        total_sum += num
        count += 1

    return total_sum / count

# Testing the function
my_list = [1, 2, 3, 4, 5]
result = mean(my_list)
print("mean value is", result)
```

```python
# 4. bölüm 3. konu son soru
def range_of_list(int_list):
    return max(int_list) - min(int_list)

my_list = [1, 2, 3, 4, 5]
result = range_of_list(my_list)
print("The range of the list is", result)
```
