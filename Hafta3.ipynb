Lütfen kullanıcıdan bir dize isteyen ve ardından ortada kelime bulunan * karakterlerinden oluşan bir çerçeve yazdıran bir program yazın. Çerçevenin genişliği 30 karakter olmalıdır. Giriş dizesinin her zaman çerçevenin içine sığacağını varsayabilirsiniz.

Giriş dizesinin uzunluğu tek sayıysa, kelimeyi iki olası merkez konumundan herhangi birine yazdırabilirsiniz

```python
word=input("Please type a sentence or word")
print("*"*30)
sayi=(28-len(word))//2#+1 e gerek yok çünkü zaten boşluk kadar eklenip yazılacak
word="*"+"!"*sayi+word+"!"*(28 - len(word) - sayi)+"*"
print(word)
print("*"*30)
```

Hata: (28 - len(word)) / 2 + 1 → Bölme (/) işlemi float türü döndürür, ancak dizelerde indeksleme tam sayı (int) gerektirir.
(28 - len(word) - sayi) geriye kalan kadar döndürür
