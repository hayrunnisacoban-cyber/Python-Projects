```python
def greatest_number(sayi1,sayi2,sayi3):
  if sayi1>sayi2 and sayi1>sayi3:
    return sayi1
  elif sayi2>sayi1 and sayi2>sayi3:
    return sayi2
  else:
    return sayi3

greatest_number(6,3,8)
```

```python
#bir cümlenin ilk ikinci ve son cümlesini veren fonksiyon
sentence=input("Please type a sentence")
def first_word(sentence):
    return sentence.split()[0]#split fonksiyonu olmadan yapılması zor

def second_word(sentence):
    return sentence.split()[1]

def last_word(sentence):
    return sentence.split()[-1]
sentence = "it was a dark and stormy python"
print(first_word(sentence))  # it
print(second_word(sentence))  # was
print(last_word(sentence))  # python


```

```python
def same_chars(sentence,index1,index2):
  if sentence[index1]==sentence[index2]:
    return True
  else:
    return False
sentence=input("Please type a sentence")
same_chars("programmer",2,5)
```

Lütfen kullanıcıdan ekleme ve çıkarma arasında seçim yapmasını isteyen bir program yazın. Seçime bağlı olarak program bir listeye bir öğe ekler veya listeden bir öğe çıkarır. Eklenen öğe her zaman listedeki son öğeden bir fazla olmalıdır. Eklenecek ilk öğe 1 olmalıdır.

Liste başlangıçta ve her işlemden sonra yazdırılır. Aşağıdaki örnek yürütmeye bir göz atın:

```python
number_list=[]
sayilar=1
option1=""
while option1 !="x":
  option1=input("a(d)d, (r)emove or e(x)it")
  if option1=="d":
    number_list.append(sayilar)
    print(f"The list is now: {number_list}")
    sayilar +=1
  elif option1=="r":
    number_list.pop()
    print("The list is now:", number_list)
  print(f"Current list: {number_list}")
print("Bye")
```

```python
number_list = []  # List to store user inputs

while True:
    new_item = int(input("New item: "))  # Get input from user

    if new_item == 0:  # Exit condition
        break

    number_list.append(new_item)  # Add item to the list

    print("The list now:", number_list)  # Print in original order
    print("The list in order:", sorted(number_list))  # Print sorted list

print("Bye!")
```
